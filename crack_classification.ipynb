```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import pandas as pd
```

# Filtering out corrupted images(if any present)


```python
num_skipped = 0
for folder_name in ("cracked", "uncracked"):
    folder_path = os.path.join("train", folder_name)
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            fobj = open(fpath, "rb")
            is_jfif = tf.compat.as_bytes("JFIF") in fobj.peek(10)
        finally:
            fobj.close()

        if not is_jfif:
            num_skipped += 1
            # Delete corrupted image
            os.remove(fpath)

print("Deleted %d images" % num_skipped)
```

> Dataset does no contain any corrupted images, hence no image has been deleted

# Generating the dataset

```python
image_size = (256, 256)
batch_size = 32

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "train",
    validation_split=0.2,
    subset="training",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "train",
    validation_split=0.2,
    subset="validation",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)
```

# Data Augmentation

> Using Data Augmentation to increase the dataset

> Here I used Horizontal flip and rotation of 0.1 to increase the size of dataset

```python
data_augmentation = keras.Sequential(
    [
        layers.experimental.preprocessing.RandomFlip("horizontal"),
        layers.experimental.preprocessing.RandomRotation(0.1)
    ]
)
```

# Configuring the dataset for performance

> Let's make sure to use buffered prefetching so we can yield data from disk without having I/O becoming blocking:

```python
train_ds = train_ds.prefetch(buffer_size=32)
val_ds = val_ds.prefetch(buffer_size=32)
```

# Defining Models Architectures

> # Model 1

```python
def make_model(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)
    # Image augmentation block
    x = data_augmentation(inputs)

    # Entry block
    x = layers.experimental.preprocessing.Rescaling(1.0 / 255)(x)
    x = layers.Conv2D(32, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.Conv2D(64, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    previous_block_activation = x  # Set aside residual

    for size in [128, 256, 512, 728]:
        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(size, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(size, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.MaxPooling2D(3, strides=2, padding="same")(x)

        # Project residual
        residual = layers.Conv2D(size, 1, strides=2, padding="same")(
            previous_block_activation
        )
        x = layers.add([x, residual])  # Add back residual
        previous_block_activation = x  # Set aside next residual

    x = layers.SeparableConv2D(1024, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.GlobalAveragePooling2D()(x)
    if num_classes == 2:
        activation = "sigmoid"
        units = 1
    else:
        activation = "softmax"
        units = num_classes

    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(units, activation=activation)(x)
    return keras.Model(inputs, outputs)

model1 = make_model(input_shape=image_size + (3,), num_classes=2)
```

> # Model 2 (InceptionV3 model)

> **This model uses pre trained InceptionV3, adding to it a average pooling layer,dropout layer and a dense layer**

```python
base_model = tf.keras.applications.InceptionV3(
    include_top=False,
    weights="imagenet",
    input_shape=(256, 256, 3),
)

# trainn the base_model
#base_model.trainable = False
base_model.trainable = True
# Create new model on top
inputs = keras.Input(shape=(150, 150, 3))
x = data_augmentation(inputs)  # Apply random data augmentation

# Pre-trained Xception weights requires that input be normalized
# from (0, 255) to a range (-1., +1.), the normalization layer
# does the following, outputs = (inputs - mean) / sqrt(var)
norm_layer = keras.layers.experimental.preprocessing.Normalization()
mean = np.array([127.5] * 3)
var = mean ** 2
# Scale inputs to [-1, +1]
x = norm_layer(x)
norm_layer.set_weights([mean, var])

# The base model contains batchnorm layers. We want to keep them in inference mode
# when we unfreeze the base model for fine-tuning, so we make sure that the
# base_model is running in inference mode here.
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(1)(x)
model = keras.Model(inputs, outputs)


model.summary()
```

# Training on above defined models

> # For model 1

```python
epochs = 40

callbacks = [
    keras.callbacks.ModelCheckpoint("save_at_{epoch}.h5"),
]
model1.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="binary_crossentropy",
    metrics=["accuracy"],
)
model1.fit(
    train_ds, epochs=epochs, callbacks=callbacks, validation_data=val_ds,
)
```

## Training For Model 3

```python
base_model = tf.keras.applications.InceptionV3(
    include_top=False,
    weights="imagenet",
    input_shape=(256, 256, 3),
)

# training the base_model
base_model.trainable = True

# Create new model on top
inputs = keras.Input(shape=(150, 150, 3))
x = data_augmentation(inputs)  # Appling data augmentation

# Pre-trained Iception weights requires that input be normalized
# from (0, 255) to a range (-1., +1.), the normalization layer
# does the following, outputs = (inputs - mean) / sqrt(var)
norm_layer = keras.layers.experimental.preprocessing.Normalization()
mean = np.array([127.5] * 3)
var = mean ** 2
# Scale inputs to [-1, +1]
x = norm_layer(x)
norm_layer.set_weights([mean, var])

x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(1)(x)
model = keras.Model(inputs, outputs)


model.summary()


model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss=keras.losses.BinaryCrossentropy(from_logits=True),
    metrics=[keras.metrics.BinaryAccuracy()],
)

epochs = 7
history1 = model.fit(train_ds, epochs=epochs, validation_data=val_ds)
```

```python
import matplotlib.pyplot as plt
accuracy_train = history1.history['binary_accuracy']
accuracy_val = history1.history['val_binary_accuracy']
plt.figure(figsize = (7,7))
plt.plot(accuracy_train,color = 'red')
plt.plot(accuracy_val,color = 'black')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend('right',labels = ('training','validation'))
plt.show()
```

### So this should work fine as after this , the accuracy on validaton set is almost same and on training set increases which can become a case of overfitting on training set.
### So I will use this as the final model for prediction on test data

# Predictions on test data

## Predicting on test data using model 2

```python
score = []
filename = []
for im in l:
    img = keras.preprocessing.image.load_img("test/"+ str(im), target_size=image_size)    
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create batch axis

    predictions = model.predict(img_array)
    filename.append(str(im))
    if (predictions[0][0]>0):
        score.append("uncracked")
    else:
        score.append("cracked")
    # score>0 -> not cracked
    # score <=0 ->cracked
```

```python
submission = pd.DataFrame((filename,score))
submission = submission.T
submission = submission.rename(columns={0: "filename", 1: "class"})
#submission = submission.set_index('filename')
submission.head()
```
